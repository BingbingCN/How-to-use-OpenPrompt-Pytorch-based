```python
import random
import pandas as pd
from IPython.display import display, HTML
def display_random_examples(dataset=super_dataset['train'], num_examples=5):
    assert num_examples < len(dataset)

    random_picks = []
    for i in range(num_examples):
        random_pick = random.randint(0,len(dataset)-1)
        random_picks.append(random_pick)

    df = pd.DataFrame(dataset[random_picks])
    display(HTML(df.to_html()))

display_random_examples(super_dataset['train'], 5)

```
